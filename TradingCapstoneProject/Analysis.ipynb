```python
data['Adj. Close Lag-1'] = data['Adj. Close'] - data['Adj. Close'].shift(1)
```

```python
data['Adj. Close Lag-1'].plot()
plt.show()
```

```python
def plot_lag(lag):
    for day in range(1, lag + 1):
        plt.plot(data['Adj. Close'] - data['Adj. Close'].shift(day))
        plt.xlabel("lag days: " + str(day))
        plt.ylabel("%s days difference" % day)
        plt.show()
```

```python
plot_lag(5)
```

```python
def plot_lag_dist(lag):
    for day in range(1, lag + 1):
        sns.distplot(data['Adj. Close'] - data['Adj. Close'].shift(day).fillna(0))
        plt.xlabel("lag days: " + str(day) + "dist")
        plt.show()
```

```python
plot_lag_dist(5)
```
